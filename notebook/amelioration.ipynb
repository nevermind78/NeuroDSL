```julia
using Base: ReentrantLock

# Définition locale (dans Main) du pool et des fonctions
const LOCAL_BUFFER_POOL = Dict{DataType, Vector{Any}}()
const LOCAL_BUFFER_POOL_LOCK = ReentrantLock()
const LOCAL_POOL_MAX_PER_TYPE = 16

function local_acquire_buffer!(::Type{T}, dims::Tuple) where T <: AbstractArray
    lock(LOCAL_BUFFER_POOL_LOCK) do
        key = T
        if haskey(LOCAL_BUFFER_POOL, key) && !isempty(LOCAL_BUFFER_POOL[key])
            buf = pop!(LOCAL_BUFFER_POOL[key])
            if size(buf) == dims
                return buf
            end
        end
    end
    return T(undef, dims...)
end

function local_release_buffer!(buf::AbstractArray)
    lock(LOCAL_BUFFER_POOL_LOCK) do
        key = typeof(buf)
        if !haskey(LOCAL_BUFFER_POOL, key)
            LOCAL_BUFFER_POOL[key] = []
        end
        if length(LOCAL_BUFFER_POOL[key]) < LOCAL_POOL_MAX_PER_TYPE
            push!(LOCAL_BUFFER_POOL[key], buf)
        end
    end
end

println("✅ Fonctions locales de buffer pool prêtes.")
```

```julia
# Redéfinition de _buf! pour utiliser le pool local
function NeuroDSL._buf!(ctx, key, proto)
    haskey(ctx, key) && return ctx[key]
    buf = local_acquire_buffer!(typeof(proto), size(proto))
    ctx[key] = buf
    return buf
end

function NeuroDSL._buf_zeros!(ctx, key, dev, dims)
    haskey(ctx, key) && return ctx[key]
    buf = local_acquire_buffer!(Array{Float32}, dims)  # Par défaut Array{Float32}
    fill!(buf, 0f0)
    ctx[key] = buf
    return buf
end

println("✅ _buf! et _buf_zeros! redéfinies avec le pool local.")
```

```julia
using NeuroDSL, Random, Base.Threads

# S'assurer que le masque causal utilise bien le verrou
NeuroDSL.causal_mask_cached  # doit être la version thread-safe

function thread_task(id)
    dev = NeuroDSL.Backend.CPUDevice()
    g = NeuroDSL.NeuroGraph(namespace=Symbol(:thread_, id), device=dev)
    NeuroDSL.set!(g, :X,     randn(Float32, 16, 16))
    NeuroDSL.set!(g, :gamma, ones(Float32, 16); is_param=true)
    NeuroDSL.set!(g, :Z,     zeros(Float32, 16, 16); atom_type=NeuroDSL.Datom)
    NeuroDSL.@addrules g Symbol(:thread_, id) begin
        H = rmsnorm(X, gamma)
        L = mse_loss(H, Z)
    end
    for _ in 1:10
        NeuroDSL.invalidate_all!(g)
        ctx = NeuroDSL.CtxStore()
        NeuroDSL.demand!(g, :L; ctx_store=ctx)
        NeuroDSL.backward_graph!(g, :L; ctx_store=ctx)
    end
    println("Thread $id terminé.")
end

println("Lancement de 4 threads...")
tasks = [Threads.@spawn thread_task(i) for i in 1:4]
for t in tasks; wait(t); end
println("✅ Test multi‑thread terminé sans erreur.")
```

```julia
using NeuroDSL

dev = NeuroDSL.Backend.CPUDevice()
g_vit = NeuroDSL.NeuroGraph(namespace=:vit_full, device=dev)

# Dimensions
batch = 1
num_patches = 64
patch_dim = 48   # 4*4*3
dim = 128
heads = 4
hidden_dim = 256
depth = 4

# --- Patch embedding (entrée aplatie 2D) ---
NeuroDSL.set!(g_vit, :patches_flat, randn(Float32, num_patches, patch_dim))  # (64,48)
NeuroDSL.set!(g_vit, :proj_W, randn(Float32, patch_dim, dim); is_param=true)  # (48,128)
NeuroDSL.set!(g_vit, :proj_b, zeros(Float32, dim); is_param=true)
NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(:emb, [:patches_flat, :proj_W], :matmul; namespace=:vit_full))
NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(:emb_b, [:emb, :proj_b], :add; namespace=:vit_full))

# --- Positional embedding ---
NeuroDSL.set!(g_vit, :pos_emb, randn(Float32, num_patches, dim); is_param=true)
NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(:x0, [:emb_b, :pos_emb], :add; namespace=:vit_full))

# --- Blocs Transformer (tous dans le même namespace) ---
current = :x0
for d in 1:depth
    blk = Symbol(:blk, d)
    # RMSNorm 1
    gamma1 = Symbol(blk, :_norm1_gamma)
    NeuroDSL.set!(g_vit, gamma1, ones(Float32, dim); is_param=true)
    h1 = Symbol(blk, :_h1)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(h1, [current, gamma1], :rmsnorm; namespace=:vit_full))
    # Attention multi‑têtes simplifiée (Q, K, V, Output)
    qw = Symbol(blk, :_qW); kw = Symbol(blk, :_kW); vw = Symbol(blk, :_vW); ow = Symbol(blk, :_oW)
    NeuroDSL.set!(g_vit, qw, randn(Float32, dim, dim); is_param=true)
    NeuroDSL.set!(g_vit, kw, randn(Float32, dim, dim); is_param=true)
    NeuroDSL.set!(g_vit, vw, randn(Float32, dim, dim); is_param=true)
    NeuroDSL.set!(g_vit, ow, randn(Float32, dim, dim); is_param=true)
    q = Symbol(blk, :_Q); k = Symbol(blk, :_K); v = Symbol(blk, :_V)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(q, [h1, qw], :matmul; namespace=:vit_full))
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(k, [h1, kw], :matmul; namespace=:vit_full))
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(v, [h1, vw], :matmul; namespace=:vit_full))
    # Scores = Q*K' / sqrt(dim)
    scores = Symbol(blk, :_scores)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(scores, [q, k], :matmul; attrs=Dict(:trans_b=>true), namespace=:vit_full))
    # Scale mask
    scaled = Symbol(blk, :_scaled)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(scaled, [scores], :scale_mask; attrs=Dict(:d_head=>(dim÷heads)), namespace=:vit_full))
    # Softmax
    attn_w = Symbol(blk, :_attn_w)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(attn_w, [scaled], :softmax; namespace=:vit_full))
    # Attention output = attn_w * V
    attn_out = Symbol(blk, :_attn_out)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(attn_out, [attn_w, v], :matmul; namespace=:vit_full))
    # Output projection
    attn_proj = Symbol(blk, :_attn_proj)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(attn_proj, [attn_out, ow], :matmul; namespace=:vit_full))
    # Résiduel 1
    res1 = Symbol(blk, :_res1)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(res1, [current, attn_proj], :add; namespace=:vit_full))
    # RMSNorm 2
    gamma2 = Symbol(blk, :_norm2_gamma)
    NeuroDSL.set!(g_vit, gamma2, ones(Float32, dim); is_param=true)
    h2 = Symbol(blk, :_h2)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(h2, [res1, gamma2], :rmsnorm; namespace=:vit_full))
    # MLP (2 couches)
    w1 = Symbol(blk, :_mlp_w1); w2 = Symbol(blk, :_mlp_w2)
    NeuroDSL.set!(g_vit, w1, randn(Float32, dim, hidden_dim); is_param=true)
    NeuroDSL.set!(g_vit, w2, randn(Float32, hidden_dim, dim); is_param=true)
    mlp1 = Symbol(blk, :_mlp1)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(mlp1, [h2, w1], :matmul; namespace=:vit_full))
    mlp2 = Symbol(blk, :_mlp2)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(mlp2, [mlp1, w2], :matmul; namespace=:vit_full))
    # Résiduel 2
    res2 = Symbol(blk, :_res2)
    NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(res2, [res1, mlp2], :add; namespace=:vit_full))
    current = res2
end

# --- Tête de classification ---
NeuroDSL.set!(g_vit, :cls_W, randn(Float32, dim, 10); is_param=true)
NeuroDSL.set!(g_vit, :cls_b, zeros(Float32, 10); is_param=true)
NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(:logits_pre, [current, :cls_W], :matmul; namespace=:vit_full))
NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(:logits, [:logits_pre, :cls_b], :add; namespace=:vit_full))

# Cible et loss
NeuroDSL.set!(g_vit, :target, zeros(Float32, num_patches, 10); atom_type=NeuroDSL.Datom)
NeuroDSL.addrule!(g_vit, NeuroDSL.GraphRule(:loss, [:logits, :target], :mse_loss; namespace=:vit_full))

println("✅ ViT unifié construit avec $(length(g_vit.nodes[:vit_full])) nœuds et $(length(g_vit.rules[:vit_full])) règles")
```

```julia
function NeuroDSL.backward_graph_sparse!(g::NeuroGraph, loss_sym::Symbol;
                                         ctx_store::CtxStore = CtxStore(),
                                         namespace::Symbol = g.active_ns)
    ns = namespace

    # 1. Réinitialiser tous les gradients
    for (_, nd) in g.nodes[ns]
        nd.gradient = nothing
        nd.backwarded = false
    end

    # 2. Initialiser le gradient de la loss
    ln = g.nodes[ns][loss_sym]
    @assert length(ln.value) == 1 "Loss must be scalar"
    ln.gradient = NeuroDSL.Backend.ones32(g.device, size(ln.value)...)
    ln.backwarded = false

    # 3. Ordre inverse
    order = NeuroDSL.topo_order!(g; namespace = ns)
    reverse_order = reverse(order)

    # 4. Compteur de références backward
    backward_refs = Dict{Symbol, Int}()
    for out_sym in reverse_order
        haskey(g.rules[ns], out_sym) || continue
        rule = g.rules[ns][out_sym]
        for in_sym in rule.inputs
            backward_refs[in_sym] = get(backward_refs, in_sym, 0) + 1
        end
    end

    # 5. Parcours backward
    for out_sym in reverse_order
        haskey(g.rules[ns], out_sym) || continue
        rule = g.rules[ns][out_sym]
        nd_out = g.nodes[ns][out_sym]

        nd_out.gradient === nothing && continue
        !haskey(NeuroDSL.GRAD_RULES, rule.op) && continue

        # Contexte forward
        ctx = get(ctx_store, out_sym, nothing)
        if ctx === nothing
            ctx_tmp = NeuroDSL.CtxStore()
            NeuroDSL.execute_rule!(g, rule; ctx_store = ctx_tmp, namespace = ns)
            ctx = get(ctx_tmp, out_sym, Dict{Symbol, Any}())
        end

        inputs_vals = [g.nodes[ns][s].value for s in rule.inputs]
        grads = NeuroDSL.GRAD_RULES[rule.op](g.device, nd_out.gradient, ctx, inputs_vals)

        for (i, in_sym) in enumerate(rule.inputs)
            in_nd = g.nodes[ns][in_sym]

            if in_nd.is_param
                NeuroDSL.accum_grad!(in_nd, grads[i])
            else
                in_nd.gradient = grads[i]
            end
            in_nd.backwarded = true

            # Décrémenter le compteur
            backward_refs[in_sym] -= 1

            # Libérer la valeur SEULEMENT si le nœud est recalculable (a une règle)
            # Les entrées (sans règle) ne doivent jamais être libérées.
            if backward_refs[in_sym] == 0 &&
               !in_nd.is_param &&
               haskey(g.rules[ns], in_sym)
                in_nd.value = nothing
                in_nd.valid = false
            end
        end

        delete!(ctx_store, out_sym)
        nd_out.backwarded = true

        if !nd_out.is_param
            nd_out.gradient = nothing
        end
    end

    # 6. Nettoyage final
    for (_, nd) in g.nodes[ns]
        if !nd.is_param
            nd.gradient = nothing
        end
    end

    return g
end
```

```julia
# Forward
NeuroDSL.invalidate_all!(g_vit)
ctx = NeuroDSL.CtxStore()
NeuroDSL.demand!(g_vit, :loss; ctx_store=ctx)

# Backward sparse (avec libération anticipée)
NeuroDSL.backward_graph_sparse!(g_vit, :loss; ctx_store=ctx)

# Vérifier que les activations ont bien été libérées
freed = count(sym -> g_vit.nodes[:vit_full][sym].value === nothing,
              keys(g_vit.nodes[:vit_full]))
println("Activations libérées après backward sparse : $freed / $(length(g_vit.nodes[:vit_full]))")
```

```julia
using NeuroDSL

ns = :vit_full

# Sauvegarder TOUS les nœuds qui n'ont pas de règle (entrées + paramètres)
entrees_et_params = Dict{Symbol, AbstractArray}()
for (sym, nd) in g_vit.nodes[ns]
    if !haskey(g_vit.rules[ns], sym) && nd.value !== nothing
        entrees_et_params[sym] = copy(nd.value)
    end
end

# Fonction de réinitialisation RESTAURANT tout ce qui n'a pas de règle
function reset_graph!(g, ns, saved)
    NeuroDSL.invalidate_all!(g)
    for (sym, val) in saved
        g.nodes[ns][sym].value = val
        g.nodes[ns][sym].valid = true
    end
end

# --- Backward standard ---
reset_graph!(g_vit, ns, entrees_et_params)
ctx_std = NeuroDSL.CtxStore()
NeuroDSL.demand!(g_vit, :loss; ctx_store=ctx_std)
mem_std = @allocated NeuroDSL.backward_graph!(g_vit, :loss; ctx_store=ctx_std)
println("Allocations standard : $(round(mem_std/1024^2, digits=2)) MiB")

# --- Backward sparse (avec libération anticipée) ---
reset_graph!(g_vit, ns, entrees_et_params)
ctx_sparse = NeuroDSL.CtxStore()
NeuroDSL.demand!(g_vit, :loss; ctx_store=ctx_sparse)
mem_sparse = @allocated NeuroDSL.backward_graph_sparse!(g_vit, :loss; ctx_store=ctx_sparse)
println("Allocations sparse  : $(round(mem_sparse/1024^2, digits=2)) MiB")

gain = round(100 * (1 - mem_sparse / max(1, mem_std)), digits=1)
println("\n✅ Réduction mémoire : $gain%")
```

```julia
using NeuroDSL

ns = :vit_full

# Créer un schedule de checkpointing (1 checkpoint tous les 2 nœuds quantiques)
cd = NeuroDSL.CheckpointData(every = 2)
schedule = NeuroDSL.CheckpointSchedule(g_vit, cd; namespace = ns)

# --- Backward standard (référence) ---
reset_graph!(g_vit, ns, entrees_et_params)
ctx_std = NeuroDSL.CtxStore()
NeuroDSL.demand!(g_vit, :loss; ctx_store = ctx_std)
NeuroDSL.backward_graph!(g_vit, :loss; ctx_store = ctx_std)
grads_std = Dict(sym => copy(nd.gradient) for (sym, nd) in g_vit.nodes[ns] if nd.is_param && nd.gradient !== nothing)

# --- Backward checkpointé ---
reset_graph!(g_vit, ns, entrees_et_params)
ctx_chk = NeuroDSL.CtxStore()
NeuroDSL.forward_with_checkpointing!(g_vit, :loss, ctx_chk, schedule; namespace = ns)
NeuroDSL.backward_with_checkpointing!(g_vit, :loss; ctx_store = ctx_chk, schedule = schedule, namespace = ns)

# Comparaison
println("=== Comparaison standard vs checkpointing ===")
max_diff = 0.0f0
for sym in keys(grads_std)
    gs = grads_std[sym]
    nd = g_vit.nodes[ns][sym]
    gp = nd.gradient
    if gp === nothing
        println("❌ $sym : absent dans le backward checkpointé")
        continue
    end
    gp = copy(gp)
    diff = norm(gs .- gp) / max(1f-8, norm(gs))
    max_diff = max(max_diff, diff)
    status = diff < 1e-4 ? "✅" : "❌"
    @printf "%s %-30s  diff relative = %.2e\n" status sym diff
end
println("Différence relative maximale : $(max_diff)")
```

```julia
function NeuroDSL.backward_graph_sparse!(g::NeuroGraph, loss_sym::Symbol;
                                         ctx_store::CtxStore = CtxStore(),
                                         namespace::Symbol = g.active_ns,
                                         recompute::Bool = true)
    ns = namespace

    # 1. Réinitialiser les gradients
    for (_, nd) in g.nodes[ns]
        nd.gradient = nothing
        nd.backwarded = false
    end

    # 2. Gradient de la loss
    ln = g.nodes[ns][loss_sym]
    @assert length(ln.value) == 1 "Loss must be scalar"
    ln.gradient = NeuroDSL.Backend.ones32(g.device, size(ln.value)...)
    ln.backwarded = false

    # 3. Ordre inverse
    order = NeuroDSL.topo_order!(g; namespace = ns)
    reverse_order = reverse(order)

    # 4. Compteur de références backward
    backward_refs = Dict{Symbol, Int}()
    for out_sym in reverse_order
        haskey(g.rules[ns], out_sym) || continue
        for in_sym in g.rules[ns][out_sym].inputs
            backward_refs[in_sym] = get(backward_refs, in_sym, 0) + 1
        end
    end

    # 5. Parcours backward
    for out_sym in reverse_order
        haskey(g.rules[ns], out_sym) || continue
        rule = g.rules[ns][out_sym]
        nd_out = g.nodes[ns][out_sym]

        nd_out.gradient === nothing && continue
        !haskey(NeuroDSL.GRAD_RULES, rule.op) && continue

        # Recalculer les entrées manquantes (si activation libérée)
        for in_sym in rule.inputs
            in_nd = g.nodes[ns][in_sym]
            if in_nd.value === nothing
                if recompute && haskey(g.rules[ns], in_sym)
                    NeuroDSL.demand!(g, in_sym; ctx_store=ctx_store, namespace=ns)
                else
                    error("Activation manquante pour $in_sym et non recalculable")
                end
            end
        end

        # Récupérer le contexte forward
        ctx = get(ctx_store, out_sym, nothing)
        if ctx === nothing
            NeuroDSL.demand!(g, out_sym; ctx_store=ctx_store, namespace=ns)
            ctx = get(ctx_store, out_sym, nothing)
            if ctx === nothing
                ctx_tmp = NeuroDSL.CtxStore()
                NeuroDSL.execute_rule!(g, rule; ctx_store=ctx_tmp, namespace=ns)
                ctx = get(ctx_tmp, out_sym, Dict{Symbol, Any}())
            end
        end

        # Calculer les gradients
        inputs_vals = [g.nodes[ns][s].value for s in rule.inputs]
        grads = NeuroDSL.GRAD_RULES[rule.op](g.device, nd_out.gradient, ctx, inputs_vals)

        # Propager les gradients
        for (i, in_sym) in enumerate(rule.inputs)
            in_nd = g.nodes[ns][in_sym]
            if in_nd.is_param
                NeuroDSL.accum_grad!(in_nd, grads[i])
            else
                in_nd.gradient = grads[i]
            end
            in_nd.backwarded = true

            backward_refs[in_sym] -= 1

            # Libérer si plus jamais utilisé
            if backward_refs[in_sym] == 0 &&
               !in_nd.is_param &&
               haskey(g.rules[ns], in_sym)
                in_nd.value = nothing
                in_nd.valid = false
            end
        end

        delete!(ctx_store, out_sym)
        nd_out.backwarded = true

        if !nd_out.is_param
            nd_out.gradient = nothing
        end
    end

    # 6. Nettoyage final
    for (_, nd) in g.nodes[ns]
        if !nd.is_param
            nd.gradient = nothing
        end
    end

    return g
end
```

```julia
# Liste des nœuds dont la libération a produit des gradients corrects (observé précédemment)
safe_to_free = Set{Symbol}([:blk4_mlp_w1, :blk4_norm2_gamma, :blk4_mlp_w2, :cls_W, :cls_b])

function NeuroDSL.backward_graph_sparse!(g::NeuroGraph, loss_sym::Symbol;
                                         ctx_store::CtxStore = CtxStore(),
                                         namespace::Symbol = g.active_ns,
                                         free_set::Set{Symbol} = Set{Symbol}())
    ns = namespace
    for (_, nd) in g.nodes[ns]
        nd.gradient = nothing
        nd.backwarded = false
    end
    ln = g.nodes[ns][loss_sym]
    ln.gradient = NeuroDSL.Backend.ones32(g.device, size(ln.value)...)
    ln.backwarded = false

    order = NeuroDSL.topo_order!(g; namespace = ns)
    reverse_order = reverse(order)

    backward_refs = Dict{Symbol, Int}()
    for out_sym in reverse_order
        haskey(g.rules[ns], out_sym) || continue
        for in_sym in g.rules[ns][out_sym].inputs
            backward_refs[in_sym] = get(backward_refs, in_sym, 0) + 1
        end
    end

    for out_sym in reverse_order
        haskey(g.rules[ns], out_sym) || continue
        rule = g.rules[ns][out_sym]
        nd_out = g.nodes[ns][out_sym]
        nd_out.gradient === nothing && continue
        !haskey(NeuroDSL.GRAD_RULES, rule.op) && continue

        # Recalcul si nécessaire (uniquement si on n'a pas déjà libéré)
        for in_sym in rule.inputs
            in_nd = g.nodes[ns][in_sym]
            if in_nd.value === nothing && haskey(g.rules[ns], in_sym)
                NeuroDSL.demand!(g, in_sym; ctx_store=ctx_store, namespace=ns)
            end
        end

        ctx = get(ctx_store, out_sym, nothing)
        if ctx === nothing
            ctx_tmp = NeuroDSL.CtxStore()
            NeuroDSL.execute_rule!(g, rule; ctx_store=ctx_tmp, namespace=ns)
            ctx = get(ctx_tmp, out_sym, Dict{Symbol, Any}())
        end

        inputs_vals = [g.nodes[ns][s].value for s in rule.inputs]
        grads = NeuroDSL.GRAD_RULES[rule.op](g.device, nd_out.gradient, ctx, inputs_vals)

        for (i, in_sym) in enumerate(rule.inputs)
            in_nd = g.nodes[ns][in_sym]
            if in_nd.is_param
                NeuroDSL.accum_grad!(in_nd, grads[i])
            else
                in_nd.gradient = grads[i]
            end
            in_nd.backwarded = true

            backward_refs[in_sym] -= 1

            # Libération SEULEMENT pour les nœuds de l'ensemble "sûr"
            if backward_refs[in_sym] == 0 && !in_nd.is_param && haskey(g.rules[ns], in_sym) && in_sym in free_set
                in_nd.value = nothing
                in_nd.valid = false
            end
        end

        delete!(ctx_store, out_sym)
        nd_out.backwarded = true
        if !nd_out.is_param
            nd_out.gradient = nothing
        end
    end

    for (_, nd) in g.nodes[ns]
        if !nd.is_param
            nd.gradient = nothing
        end
    end
    return g
end
```

```julia
import NeuroDSL.backward_with_checkpointing!

function backward_with_checkpointing!(g::NeuroGraph, loss_sym::Symbol;
                                      ctx_store::CtxStore = CtxStore(),
                                      schedule::CheckpointSchedule,
                                      namespace::Symbol = g.active_ns)
    ns = namespace
    zero_grads!(g; namespace=ns)
    ln = node(g, loss_sym; namespace=ns)
    @assert length(ln.value)==1 "loss doit être scalaire"
    ln.gradient = Backend.ones32(g.device, size(ln.value)...)

    for out_sym in reverse(schedule.order)
        !haskey(g.rules[ns], out_sym) && continue
        rule = g.rules[ns][out_sym]
        nd_out = g.nodes[ns][out_sym]
        nd_out.gradient === nothing && continue

        !haskey(GRAD_RULES, rule.op) &&
            error("❌ Pas de règle backward pour :$(rule.op)")

        # 1. Recalculer les entrées manquantes (la nouvelle _recompute_segment!
        #    va maintenant stocker leurs contextes dans ctx_store)
        for in_sym in rule.inputs
            in_nd = get(g.nodes[ns], in_sym, nothing)
            if in_nd !== nothing && (in_nd.value === nothing || !in_nd.valid)
                _recompute_segment!(g, in_sym, schedule, ctx_store; namespace=ns)
            end
        end

        # 2. Maintenant que les entrées sont dispo ET leurs contextes stockés,
        #    on réexécute la règle de out_sym pour mettre à jour SON contexte
        execute_rule!(g, rule; ctx_store=ctx_store, namespace=ns)

        # 3. Récupération du contexte (garanti présent)
        ctx = get(ctx_store, out_sym, nothing)
        if ctx === nothing
            ctx_tmp = CtxStore()
            execute_rule!(g, rule; ctx_store=ctx_tmp, namespace=ns)
            ctx = get(ctx_tmp, out_sym, Dict{Symbol, Any}())
        end

        inputs_vals = [g.nodes[ns][s].value for s in rule.inputs]
        grads = GRAD_RULES[rule.op](g.device, nd_out.gradient, ctx, inputs_vals)

        for (i, in_sym) in enumerate(rule.inputs)
            accum_grad!(g.nodes[ns][in_sym], grads[i])
        end

        delete!(ctx_store, out_sym)
        nd_out.gradient = nothing
    end
    return g
end
```

```julia
function _recompute_segment!(g::NeuroGraph, target_sym::Symbol,
                             schedule::CheckpointSchedule,
                             ctx_store::Union{CtxStore,Nothing}=nothing;
                             namespace::Symbol = g.active_ns)
    order = schedule.order
    ns = namespace
    target_idx = findfirst(==(target_sym), order)
    target_idx === nothing && return nothing

    # Cherche le dernier checkpoint ou nœud encore valide avant la cible
    start_idx = 1
    for i in (target_idx-1):-1:1
        prev_sym = order[i]
        nd = get(g.nodes[ns], prev_sym, nothing)
        if nd !== nothing && nd.value !== nothing && nd.valid
            start_idx = i + 1
            break
        end
    end

    # Recalcule tous les nœuds de start_idx jusqu'à target_idx
    for i in start_idx:target_idx
        sym = order[i]
        nd = get(g.nodes[ns], sym, nothing)
        nd === nothing && continue
        nd.valid && continue

        # ICI : on force la réexécution complète de la règle pour que
        # le contexte soit stocké dans ctx_store
        if haskey(g.rules[ns], sym)
            rule = g.rules[ns][sym]
            # demand! appelle execute_rule! qui, elle, stocke le contexte
            demand!(g, sym; ctx_store=ctx_store, namespace=ns)
        else
            # Nœud sans règle (entrée) : ne devrait pas arriver
            continue
        end
    end

    return g.nodes[ns][target_sym].value
end
```
